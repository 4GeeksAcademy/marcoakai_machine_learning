Division del dataset en nuestra variable dependiente y las descriptoras

```python
target = "Family Medicine/General Practice Primary Care (2019)"

# Eliminamos filas donde la variable objetivo esté vacía
df = df.dropna(subset=[target])

X = df.drop(columns=[target])
y = df[target]
```

SPLITE

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=18)
X_train.shape, X_test.shape
```

Normalizacion

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```


Entrenamiento del modelo

```python
model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

Predicción del modelo

```python
y_pred = model.predict(X_test_scaled)
```

evaluacion del modelo

```python
R2 = r2_score(y_test, y_pred)
RMSE = np.sqrt(mean_squared_error(y_test, y_pred))

R2, RMSE
```

lasso model

```python
lasso = Lasso(alpha=1.0, max_iter=10000)
lasso.fit(X_train_scaled, y_train)


y_pred_lasso = lasso.predict(X_test_scaled)
```

```python
f"Error cuadrático medio: {mean_squared_error(y_test, y_pred_lasso)}"
```

```python
"Coeficiente de determinación:", r2_score(y_test, y_pred_lasso),
```
